```python
fb = pd.read_excel(r"D:\Notebooks\run only EDA\Data_Science-main\Data\facebook_web_log.xlsx")
fb.head()
```

```python
fb = fb.query("action in ('page_load','page_exit')")
fb['page_exit'] = fb.apply(lambda x: x['timestamp'] if x['action']=='page_exit' else None, axis=1)
fb['page_load'] = fb.apply(lambda x: x['timestamp'] if x['action']=='page_load' else None, axis=1)
fb['page_loadN']=fb['page_load'].fillna(method='ffill')
fb = fb.dropna(subset=['page_exit'])
```

```python
fb = fb[['user_id','page_exit','page_loadN']]
fb['sec_diff'] = (fb['page_exit']-fb['page_loadN']).dt.total_seconds()
fb
```

```python
fb.pivot_table(index='user_id',values='sec_diff',aggfunc='mean').reset_index()
```

```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
```

```python
data  = {"Income_K":[75,67,70,90,80]
         ,"Profession":["Salaried","Business","Salaried","Business","Salaried"]
        ,"Loan_Lakhs":[3,2,5,7,4]}

df = pd.DataFrame(data)
df
```

```python
le = LabelEncoder()
```

```python
df['prof_enc'] = le.fit_transform(df['Profession'])
df
```

```python
x = df[['Income_K','prof_enc']]
y = df['Loan_Lakhs']
```

```python
model = GradientBoostingRegressor(n_estimators=5)
```

```python
model.fit(x,y)
```

```python
prediction = model.predict(x)
```

```python
sum((y-prediction)**2)/len(y)
```

```python
params = {'n_estimators':range(1,200)}
grid = GridSearchCV(param_grid=params,estimator=model,cv=2, scoring='neg_mean_squared_error')
```

```python
grid.fit(x,y)
```

```python
grid.best_estimator_
```

```python
gb = grid.best_estimator_
```

```python
gb.fit(x,y)
```

```python
preodiction = gb.predict(x)
```

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(salary[['YearsExperience']],salary.Salary,test_size=0.2,random_state=False)
```

```python
y_test.shape
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test,y_test)
```
